```python
movies = movies.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies['actor_1_name'] = movies['actor_1_name'].replace(np.nan, 'null')
movies['actor_2_name'] = movies['actor_2_name'].replace(np.nan, 'null')
movies['actor_3_name'] = movies['actor_3_name'].replace(np.nan, 'null')
movies['director_name'] = movies['director_name'].replace(np.nan, 'null')
```

```python
movies
```

```python
movies['genres'] = movies['genres'].str.replace('|' , ',')
```

```python
movies
```

```python
movies['movie_title'].str.isupper().sum()
```

```python
movies['movie_title'] = movies['movie_title'].str.lower()
```

```python
movies.head()
```

```python
movies['movie_title'] = movies['movie_title'].apply(lambda x : x[:-1])
```

```python
movies.to_csv('movies.csv', index = False)
```
